```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../DATA/cluster_mpg.csv")
```

```python
df
```

```python
df.describe().transpose()
```

```python
df["origin"].value_counts()
```

```python
df_dummies = pd.get_dummies(df.drop("name", axis=1))
```

```python
df_dummies
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaled_data = scaler.fit_transform(df_dummies)
```

```python
scaled_data
```

```python
scaled_df = pd.DataFrame(scaled_data, columns=df_dummies.columns)
```

```python
scaled_df
```

```python
plt.figure(figsize=(15, 10), dpi=200)
sns.heatmap(scaled_df, cmap="magma")
```

```python
len(df)
```

```python
plt.figure(figsize=(15, 10), dpi=200)
sns.clustermap(scaled_df, cmap="magma", row_cluster=False)
```

```python
plt.figure(figsize=(15, 10), dpi=200)
sns.clustermap(scaled_df, cmap="magma", col_cluster=False)
```

```python
sns.heatmap(scaled_df.corr())
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=4)
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
plt.figure(figsize=(12, 6), dpi=200)
sns.scatterplot(data=df, x="mpg", y="horsepower", hue=cluster_labels, palette="viridis")
```

```python
# Maximum possible distance between two points (Min/Max Scaler)
np.sqrt(len(scaled_df.columns))
```

```python
# When distance_threshold = 0 (Meaning even if two points have 0 distance between them they can't be joined to form a cluster)
model = AgglomerativeClustering(n_clusters=None, distance_threshold=0)
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
# Every point becomes a cluster!
cluster_labels
```

```python
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import linkage
```

```python
linkage_matrix = linkage(model.children_)
```

```python
linkage_matrix
```

```python
plt.figure(figsize=(20,10), dpi=200)
dendro = dendrogram(linkage_matrix)
```

```python
# X-axis here refers to cluster numbers assigned by the linkage method (Not very interpretable)
plt.figure(figsize=(20,10), dpi=200)
dendro = dendrogram(linkage_matrix, truncate_mode="lastp", p=10)
```

```python
plt.figure(figsize=(20,10), dpi=200)
dendro = dendrogram(linkage_matrix, truncate_mode="level", p=5)
```
